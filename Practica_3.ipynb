## **Programa 3:Tablas de Frecuencia **


```python
# Basado el Libro estadistica para administración pagina 12
# creamos una lista vacia qeu representa la edad de 100 personas
import random
edades = []
# Hacemos un ciclo para generar 100 numeros aleatorios que representan las edades de las personas
for i in range(100):
  numero_aleatorio = random.randint(18, 65)
  edades.append(numero_aleatorio)

print(edades)
```

```python
import pandas as pd
```

```python
# Crear un dataframe de pandas con los numeros aleatorios
df = pd.DataFrame(edades, columns=['Edades'])
df.head()
```

```python
# Calcular la tabla de distribución de frecuencias
tabla_frecuencia = df['Edades'].value_counts().reset_index()
tabla_frecuencia.columns = ['Edades', 'Frecuencia']
tabla_frecuencia = tabla_frecuencia.sort_values(by='Edades')

#Imprimir la tabla de distribución de frecuencias
print("Tabla de Distribución de Frecuencias:")
tabla_frecuencia
```

```python
tabla_frecuencia["Frec Relativa"] = tabla_frecuencia["Frecuencia"] / 100
tabla_frecuencia
```

```python
tabla_frecuencia["Frec Porcentaje"] = tabla_frecuencia["Frecuencia"] / 100 * 100
tabla_frecuencia
```

```python
tabla_frecuencia["Frec acomulada"] = tabla_frecuencia["Frecuencia"].cumsum()
tabla_frecuencia
```

```python
import math
# Calcular el numero del intervalo utilizando la regla de sturgs
n = len(edades)
k = int(1 + math.log2(n))
print("Valor de n =", n, "\nNumero de intervalos =", k)
```
