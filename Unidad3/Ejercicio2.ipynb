```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 10, 20, 50])
px = np.array([0.50, 0.30, 0.15, 0.05])

esperanza = np.sum(x * px)
varianza = np.sum((x**2) * px) - esperanza**2
desviacion = np.sqrt(varianza)

print(f"E(X): {esperanza}")
print(f"Varianza: {varianza}")
print(f"Desviación: {desviacion:.2f}")

fx = np.cumsum(px)

plt.figure(figsize=(8, 4))
plt.step(np.append(x, 60), np.append(fx, 1.0), where='post', color='green')
plt.title('Función de Distribución Acumulada F(x)')
plt.xlabel('Pesos')
plt.ylabel('F(x)')
plt.grid(True, alpha=0.3)
plt.show()
```
